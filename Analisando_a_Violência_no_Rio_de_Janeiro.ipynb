## Obtenção dos Dados

O portal ISPDados é uma página de Dados Abertos do Instituto de Segurança Pública do Rio de Janeiro, onde podemos [acessar as bases de dados](http://www.ispdados.rj.gov.br/estatistica.html) de registros criminais e de atividade policial do estado do Rio de Janeiro. De acordo com o site, as estatísticas são construídas a partir dos Registros de Ocorrência (RO) lavrados nas delegacias da Secretaria de Estado de Polícia Civil do Rio de Janeiro (SEPOL), além de informações complementares de órgãos específicos da Secretaria de Estado de Polícia Militar do Rio de Janeiro (SEPM). 

As informações sobre os RO são submetidos ao controle de qualidade realizado pela Corregedoria Geral de Polícia (CGPOL) antes de serem consolidados no ISP. As estatísticas produzidas baseiam-se na data em que foi confeccionado o Registro de Ocorrência.

Por termos atributos do tipo `object` o pandas reconhece os valores em brancos como Strings, como se houvesse o caracter "espaço" neste campo, portanto, precisarão ser convertidos para o tipo numérico.

O `Pandas` possui o método `replace()` que nos permite substituir determinados elementos do _DataFrame_ por outros, então substituiremos esses "espaços" pelo valor nulo `NaN` do `numpy`, no qual o `pandas` atribui os valores nulos.

```
# Substituindo os espaços em brancos por valores Nulos 

df.replace(r" ", np.nan, inplace=True)
```

Agora só precisamos converter essas variáveis do tipo `object` para o tipo `float`.

```
# Gerando uma lista com o nome dos atributos do tipo 'object'
features_obj = df.select_dtypes('object').columns

# Transformando esses atributos para o tipo float
df[features_obj] = df[features_obj].astype('float')
```

### Crimes contra o Patrimônio - Veículos

Um dos crimes mais comuns são os relacionados a veículos. É um problema bastante recorrente nas metrópoles brasileiras e esta seção será dedicada a esse problema. Por termos dados sobre roubo, furto e recuperação de veículos, conseguiremos usar essas informações para nos situarmos sobre o resultado das ações de combate a esse tipo de crime.

Antes de entrarmos em detalhes, devemos ressaltar a diferença entre _furto_ e _roubo_, pois de acordo com o [Tribunal de Justiça do Distrito Federal e dos Territórios](https://www.tjdft.jus.br/institucional/imprensa/campanhas-e-produtos/direito-facil/edicao-semanal/furto-e-roubo):
* **Furto**: Subtração de um patrimônio sem que haja violência.
* **Roubo**: Caracterizado como subtração do patrimônio sob grave ameaça ou violência. 

Veremos os valores para o roubo de veículos.


#### **Roubo de Veículos**

```
# Verificando os valores mínimo, máximo e a média por mês do atributo 'roubo_veiculo'
roubo_min = df.roubo_veiculo.min()
roubo_max = df.roubo_veiculo.max()
roubo_med = round(df.roubo_veiculo.mean(),2)

print("Roubo de veículos (Mínimo): ", roubo_min)
print("Roubo de veículos (Máximo): ", roubo_max)
print("Roubo de veículos (Média): ", roubo_med)
```

```
fig, ax = plt.subplots(figsize=(5,3))


ax.text(0.0, 1, s=f'{round(roubo_med)}', size=60, color=LARANJA, fontweight='bold')
ax.text(0.0, 0.750, s='veículos são $\\bf{roubados}$\nem média por mês.', size=14, color=CINZA)

sns.despine(left=True, bottom=True)
ax.xaxis.set_visible(False)
ax.yaxis.set_visible(False)
plt.tight_layout()
```

## Conclusão

Mesmo se tratando de `dataset` relativamente pequeno e simples de lidar, tivemos que fazer um tratamento e limpeza de dados em alguns campos que estava de forma "incorreta" ou com valores faltantes. 

Analisando os dados sobre diversos tipos de crimes contra o patrimônio e com os dados sobre crimes contra a vida, pudemos concluir que: 

- Um pouco menos da metade dos carros perdidos em Furtos ou Roubos são recuperados.
- Alguns dados começaram a serem registrados após uma determinada data, o que não implica que este tipo de ocorrência não acontecia antes, eles poderiam estar sendo registrados em outra categoria de crime, como o caso dos roubos e furtos de bicicletas, que só começaram a ser registrados a partir de 2014.
- Os números de Roubos apresentam uma leve crescente a partir do ano de 2010, diferente dos números de Furtos que começaram a seguir uma diminuição a partir de 2010. 
- Os números de Roubos permaneceam superiores aos números de Furtos.
- Os crimes violentos foram variando com altos e baixos ao longo dos anos, mas com tendência a diminuir.

É interessante voltar a atenção a um detalhe encontrado na análise, os meses que mais se destacaram tanto em questão de roubos/furtos e crimes contra a vida, foram os meses de **Fevereiro** e **Março**, o que implica que, nesse período de tempo, ocorre a realização do [*Carnaval*](https://pt.wikipedia.org/wiki/Carnaval), onde o fluxo de turístas aumenta consideravelmente durante esse período que por sua vez, possa estar correlacionado com os altos valores que esses meses atingiram em comparação aos outros meses.

Outro detalhe importante que deve ser ressaltado é a queda súbita de crimes durante o ano de 2020. Durante esse ano, o mundo se encontrava diante de uma pandemia e no Brasil, diversos estados decretaram o regime de *lockdown*, que reduz a circulação de pessoas nas ruas durante o período em que esse regime esteja em vigor. O baixo fluxo de pessoas circulando pela cidade pode ter influenciado na quantidade de ocorrências registradas para o ano de 2020.

Em uma análise inicial e de modo geral, pudemos obter interessantes `insights` sobre a evolução da violência no Rio de Janeiro entre o período de 1991 até o começo de 2021, que ainda nos abre um leque de opções para futuras análises mais detalhadas desse `dataset`.
